# connect and import data from a postgres server

```python
import pandas as pd
from sqlalchemy import create_engine

engine=create_engine("postgresql://postgres:bleach#postgres@localhost:5433/film")

#table query
table_query=''' 
SELECT table_name FROM information_schema.tables
WHERE table_schema='public' AND table_type ='BASE TABLE';
'''

table_names=pd.read_sql(table_query,engine)['table_name'].to_list() #to_list will not work in a dataframe

dfs={}
for table in table_names:
    dfs[table]=pd.read_sql(f"select * from public.{table}",engine)

payment = dfs["payment"]
film = dfs["film"]
actor = dfs["actor"]
address = dfs["address"]
category = dfs["category"]
city = dfs["city"]
country = dfs["country"]
customer = dfs["customer"]
film_actor = dfs["film_actor"]
film_category = dfs["film_category"]
inventory = dfs["inventory"]
language = dfs["language"]
rental = dfs["rental"]
staff = dfs["staff"]
store = dfs["store"]
```

# order by columns

DataFrame.sort_values(
    by,
    axis=0,
    ascending=True,
    inplace=False,
    kind='quicksort',
    na_position='last',
    ignore_index=False,
    key=None
)

### by- not needed in series

| Parameter          | Description                                                                |
| ------------------ | -------------------------------------------------------------------------- |
| **`by`**           | Column or list of columns to sort by. **Required**.                        |
| **`axis`**         | Sort along which axis. `0` = index (rows), `1` = columns. Default is `0`.  |
| **`ascending`**    | Sort ascending vs. descending. Can be `True`/`False` or list of bools.     |
| **`inplace`**      | If `True`, modifies the original DataFrame. Default is `False`.            |
| **`kind`**         | Sorting algorithm: `'quicksort'`, `'mergesort'`, `'heapsort'`, `'stable'`. |
| **`na_position`**  | `'first'` puts NaNs at the beginning, `'last'` puts NaNs at the end.       |
| **`ignore_index`** | If `True`, the result will have a new index from 0, 1, …, n-1.             |
| **`key`**          | A function to transform the data before sorting.                           |

```python
customer.sort_values(by=['first_name','last_name'],axis=0,ascending=[False,True],inplace=False,kind='mergesort',na_position='first',ignore_index=True,key=None)
```

# count()
DataFrame.count(axis=0, numeric_only=False)

| Parameter      | Description                                                     |
| -------------- | --------------------------------------------------------------- |
| `axis=0`       | Count along **columns** (default) → gives **count per column**  |
| `axis=1`       | Count along **rows** → gives **count per row**                  |
| `numeric_only` | If `True`, includes only numeric columns (useful with `axis=0`) |

```python
# true unique
payment.drop_duplicates(subset='amount',keep=False,ignore_index=True,inplace=False).shape[0]
```

# count() - len() - df.size - df.shape
| Function  | Counts what?                               |
| --------- | ------------------------------------------ |
| `count()` | **Non-null** values only                   |

| `df.shape`| gives a tuple (no: of rows, no: ofcolumns) |
| `df.size` | **All cells** (rows × columns) |  -> size() clubed with groupby objects to know the rows with nans

# value_counts()
Series.value_counts(normalize=False, dropna=True, ascending=False)
You call it on a Series, not on a full DataFrame.

| Parameter   | Description                                              |
| ----------- | -------------------------------------------------------- |
| `normalize` | If `True`, returns **proportions** instead of raw counts |
| `dropna`    | If `False`, includes NaN in the counts                   |
| `ascending` | Sort result in **ascending order**                       |

```python
# value count for count filters
fe=payment['amount'].value_counts(normalize=False,dropna=False,ascending=True).reset_index()
fe.columns = ['amount', 'counter'] # renamed columns
fe.loc[fe['counter']==1,['amount']]
```

# set_index()
df.set_index('column_name', inplace=False, drop=True)

| Parameter | Description                                             |
| --------- | ------------------------------------------------------- |
| `inplace` | Modify original DataFrame or return new one             |
| `drop`    | If `True`, removes the column after setting it as index |

```python
fe.set_index('amount',inplace=True,drop=True)
```

# reset_index()

| Parameter   | Description                                             |
| ----------- | ------------------------------------------------------- |
| `drop=True` | Drops the index completely, instead of moving to column |
| `inplace`   | Apply directly or return a new DataFrame                |

```python
# reset_index
fe.reset_index(drop=False,inplace=True)
fe
```
